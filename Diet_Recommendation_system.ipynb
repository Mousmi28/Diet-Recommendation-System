```python
columns_to_drop = ['Patient_ID', 'Dietary_Restrictions', 'Allergies', 'Preferred_Cuisine','Blood_Pressure_mmHg','Glucose_mg/dL','Adherence_to_Diet_Plan','BMI']
df = df.drop(columns=[col for col in columns_to_drop if col in df.columns])
```

```python
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()
df['Gender'] = le.fit_transform(df['Gender'])
df['Diet_Recommendation']=le.fit_transform(df['Diet_Recommendation'])
```

```python
from sklearn.preprocessing import OneHotEncoder
import pandas as pd

# List of categorical columns to encode
categorical_cols = ['Disease_Type', 'Severity', 'Physical_Activity_Level']

# Initialize OneHotEncoder with sparse=False
encoder = OneHotEncoder(sparse_output=False, handle_unknown='ignore')  

# Fit and transform categorical columns 
encoded_array = encoder.fit_transform(df[categorical_cols])

# Convert to DataFrame with proper column names
encoded_df = pd.DataFrame(encoded_array, columns=encoder.get_feature_names_out(categorical_cols))

# Reset index to match the original DataFrame
encoded_df.index = df.index

# Concatenate encoded data with original DataFrame and drop original categorical columns
df = pd.concat([df.drop(columns=categorical_cols), encoded_df], axis=1)
```

```python
df = df.reset_index(drop=True)
```

```python
df.columns
```

```python
df
```

```python
X = df.drop(columns=['Diet_Recommendation'])   
y = df['Diet_Recommendation']   
```

```python
df.isna().sum() 
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2,random_state=32)
```

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
X_train_scaled=scaler.fit_transform(X_train)
X_test_scaled=scaler.transform(X_test)
```

```python
from sklearn.linear_model import LogisticRegression
lr = LogisticRegression()
lr.fit(X_train,y_train)
pred_lr = lr.predict(X_test)
lr_acc = lr.score(X_test,y_test)
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test,pred_lr))
```

```python
from sklearn.tree import DecisionTreeClassifier
dc = DecisionTreeClassifier()
dc.fit(X_train,y_train)
pred_dc = dc.predict(X_test)
dc_acc = dc.score(X_test,y_test)
print(classification_report(y_test,pred_dc))
```

```python
from sklearn.naive_bayes import MultinomialNB
nb = MultinomialNB()
nb.fit(X_train,y_train)
pred_nb = nb.predict(X_test)
nb_acc = nb.score(X_test,y_test)
print(classification_report(y_test,pred_nb))
```

```python
from sklearn.ensemble import AdaBoostClassifier
ac = AdaBoostClassifier()
ac.fit(X_train,y_train)
pred_ac = ac.predict(X_test)
ac_acc = ac.score(X_test,y_test)
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test,pred_ac))
```

```python
from sklearn.ensemble import RandomForestClassifier
rfc = RandomForestClassifier()
rfc.fit(X_train,y_train)
pred_rfc = rfc.predict(X_test)
rfc_acc = rfc.score(X_test,y_test)
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test,pred_rfc))
```

```python
models = pd.DataFrame({
    'Model' : ['Logistic Regression','Decision Tree Classifier', 'MultinomialNB','AdaBoostClassifier','Random Forest Classifier'],
    'Accuracy Score' : [lr_acc, dc_acc, nb_acc,ac_acc, rfc_acc]})
     
    
models.sort_values(by = 'Accuracy Score', ascending = False)
```

```python
plt.figure(figsize=(12, 6))
sns.barplot(data=models.sort_values(by = 'Accuracy Score', ascending = False), x='Model', y='Accuracy Score')
plt.title('Accuracy Scores of Different Models')
plt.ylabel('Accuracy Score')
plt.show()
```
